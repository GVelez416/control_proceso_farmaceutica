```python
#Creacion del data base (C)
def createDB():
  conn = sql.connect('Proyecto.db')
  conn.commit()
  conn.close()
  print('Base de datos creada')

#Creacion de la tabla (C)
def tableDB():
  conn = sql.connect('Proyecto.db') # Si no existe la crea
  datos_df.to_sql('productos', conn, if_exists='replace', index = False)
  conn.commit()
  conn.close()
  print('Tabla creada')

#Ver la tabla (R)
def viewtable():
  conn = sql.connect('Proyecto.db')
  cursor = conn.cursor()
  query = f"SELECT * FROM productos;"
  cursor.execute(query)
  reading = cursor.fetchall()
  conn.commit()
  print('Se realizó una lectura sobre la tabla')
  print(pd.DataFrame(reading))
  conn.close()

#Modificar la tabla (U)
def updatefield():
  conn = sql.connect('Proyecto.db')
  cursor = conn.cursor()
  query = f"UPDATE productos SET estado = 'Liberado al mercado' WHERE estado like 'Liberado'; "
  cursor.execute(query)
  conn.commit()
  conn.close()
  print('Se actualizan los registros liberados')

#borrar datos de la tabla (D)
def delete():
  conn = sql.connect('Proyecto.db')
  cursor = conn.cursor()
  query = f"DELETE FROM productos WHERE cliente like 'BCR';"
  cursor.execute(query)
  conn.commit()
  conn.close()
  print('Se borra un registro')

#principal
if __name__ == '__main__':
#  createDB()
#  tableDB()
#  viewtable()
#  updatefield()
#  delete()
```
